# Data Visualization with Seaborn

```python
import seaborn as sns
```

```python
## Basic Plotting with Seaborn  
tips = sns.load_dataset('tips')

print(tips)
```

```python
# Create a scatter plot 
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y= 'tip', data= tips)
plt.title("Scatter PLot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot 

sns.lineplot(x='size', y= 'total_bill', data= tips) 
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plot 
## Bar Plot 

sns.barplot(x='day', y= 'total_bill', data = tips)
plt.title("Bar PLot of total bill by Day")
plt.show()
```

```python
## Box Plot 

sns.boxplot(x= 'day', y= 'total_bill', data= tips)
```

```python
## Voilin PLot 

sns.violinplot(x= 'day', y= 'total_bill', data= tips)
```

```python
## Histograms

sns.histplot(tips['total_bill'], bins= 10, kde= True)
```

```python
## KDE Plot 
sns.kdeplot(tips['total_bill'], fill= True)
```

```python
## Pair Plot 
sns.pairplot(tips)
```

```python
## Heatmap 

corr = tips[['total_bill', 'tip', 'size']].corr() 
corr
```

```python
sns.heatmap(corr, annot= True, cmap= 'coolwarm')
```

```python
import pandas as pd 
sales_df = pd.read_csv('Datasets/sales_data.csv')
sales_df.head(5)
```

```python
## Plot total sales by product 

plt.figure(figsize=(10, 6))
sns.barplot(x= 'Sub_Category', y= 'Revenue', data= sales_df , estimator= sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Plot total sales by Country	 

plt.figure(figsize=(10, 6))
sns.barplot(x= 'Country', y= 'Revenue', data= sales_df , estimator= sum)
plt.title('Total Sales by Country')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```
